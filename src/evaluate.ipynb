```python
print("Evaluating metrics for linear regression:")

mse = mean_squared_error(y_value, y_pred_lr)
mae = mean_absolute_error(y_value, y_pred_lr)
rmse = np.sqrt(mse)
r2 = r2_score(np.expm1(y_value), y_pred_lr)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("R squared score:", r2)
```
